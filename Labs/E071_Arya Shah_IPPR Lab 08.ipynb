# Aim: To get the Skeleton of the given image using structuring element of two sizes

```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from skimage.color import rgb2gray, rgba2rgb
import cv2
```

```python
image = io.imread('letter_j.JPG')
```

```python
image.shape
```

```python
#Convert rgb image to gray image
image=rgb2gray(image)
```

```python
sh=image.shape
sh
```

```python
image = image*255
```

```python
image[60,50]
```

```python
rows = sh[0]
cols = sh[1]
```

```python
th=180
image1 = image.copy()
rows = sh[0]
cols = sh[1]
for r in range(rows):
    for c in range(cols):
        if image1[r][c] > th:
              image1[r][c] = 255
        else: 
            image1[r][c]=0
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(image, cmap = 'gray')
plt.title('Original Image')
plt.subplot(1,2,2)
plt.imshow(image1, cmap = 'gray')
plt.title('Thresholded Image')
```

```python
img_sk = np.zeros((rows,cols))
img_sk
```

```python
se = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
se
```

```python
img_open = cv2.morphologyEx(image1,cv2.MORPH_OPEN,se) 
img_temp = cv2.subtract(image1,img_open) # subtract open image from eroded image
img_erode = cv2.erode(image1,se)
```

```python
# Image union(bitwise-OR)
for r in range(rows):
    for c in range(cols):
        if img_sk[r,c] == 0 and img_temp[r,c] == 0:
            img_sk[r,c]=0
        else:
            img_sk[r,c] = 255
```

```python
image1 = img_erode.copy()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img_erode, cmap = 'gray')
plt.title('Eroded Image')
plt.subplot(1,2,2)
plt.imshow(img_sk, cmap = 'gray')
plt.title('Skeletonized Image')
```
