```python
import matplotlib.pyplot as plt
from skimage import data
from skimage.color import rgb2gray
```

```python
image = data.astronaut()
```

```python
sh = image.shape
sh
```

```python
plt.imshow(image)
```

```python
image_gray = rgb2gray(image)
plt.imshow(image_gray, cmap= 'gray')
```

```python
image_gray[10, 10]
```

```python
image_gray = image_gray*255
image_gray[10, 10]
```

```python
plt.imshow(image_gray, cmap= 'gray')
```

```python
image_int = image_gray.copy()
b = 3
s = int(255/((2**b)-1))
s
```

```python
num_levels = 2**b
num_levels
```

```python
sh = image_int.shape
rows = sh[0]
cols = sh[1]
sh
```

```python
for r in range(rows):
  for c in range(cols):
    temp = image_int[r][c]
    for i in range(num_levels):
      if temp >= i*s and temp < (i+1)*s:
        image_int[r][c] = i*s
        break
```

```python
image_int1 = image_gray.copy()
b1 = 2
s1 = int(255/((2**b1)-1))
s1

num_levels1 = 2**b1
num_levels1

sh1 = image_int1.shape
rows1 = sh1[0]
cols1 = sh1[1]
sh1

for r in range(rows1):
  for c in range(cols1):
    temp1 = image_int1[r][c]
    for i in range(num_levels1):
      if temp1 >= i*s1 and temp1 < (i+1)*s1:
        image_int1[r][c] = i*s1
        break
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(image_gray, cmap= 'gray')
plt.title('8-bit image')

plt.subplot(1,3,2)
plt.imshow(image_int, cmap= 'gray')
plt.title('3-bit image')

plt.subplot(1,3,3)
plt.imshow(image_int1, cmap= 'gray')
plt.title('2-bit image')
```

# Conclusion

8-bit image is converted to 3-bit image using image resolution. It is observed that the modified image shows 8 gray levels. Same 8-bit image is changed to 2-bit image and it can be shown that the quality of image reduces if no of bits b is reduced. For b=1 the image can be converted to binary, whereas for higher values of b, several gray levels are merged to one level which is why the image is patchy

```python
image1 = io.imread("face.png")
plt.imshow(image1, cmap= 'gray')
```

```python
sh2 = image1.shape
sh2
rows = sh2[0]
cols = sh2[1]
```

```python
image_down = image1.copy()
ds = 20
```

```python
rowsd= int(rows/ds)
colsd= int(cols/ds)
rowsd, colsd
```

```python
image_down = image_down[0:rowsd+1, 0:colsd+1]
plt.imshow(image_down)
```

```python
rw = 0
for r in range(0, sh2[0], ds):
  cl = 0
  for c in range(0, sh2[1], ds):
    image_down[rw][cl] = image1[r][c]
    cl = cl + 1
  rw = rw + 1

plt.imshow(image_down)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,3,1)
plt.imshow(image1)
plt.title('Original')

plt.subplot(1,3,2)
plt.imshow(image_down)
plt.title('Sampling Rate 20')
```

# Conclusion

The given image is downsampled with a sampling rate of 10.

If sampling rate is changed to 20

It is observed that the image looks pixelated. This is because for the same physical size of the image, size of each pixel increase with the increase in downsampling rate.
