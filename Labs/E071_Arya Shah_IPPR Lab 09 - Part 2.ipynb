```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from skimage.color import rgb2gray, rgba2rgb
import cv2
from skimage.color import hsv2rgb
```

```python
img = cv2.imread('color_pencils.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,3,1)
plt.imshow(img_hsv[:,:,0], cmap = 'hsv')
plt.title('Hue')
plt.colorbar()
plt.subplot(1,3,2)
plt.imshow(img_hsv[:,:,1], cmap = 'gray')
plt.title('Saturation')
plt.colorbar()
plt.subplot(1,3,3)
plt.imshow(img_hsv[:,:,2], cmap = 'gray')
plt.title('Value')
plt.colorbar()
```

```python
low_th = img_hsv[:,:,0] > 160
high_th = img_hsv[:,:,0] < 180

filter_h = low_th*high_th
```

```python
img_h = img.copy()
```

```python
img_h[:,:,0] = img[:,:,0]*filter_h
img_h[:,:,1] = img[:,:,1]*filter_h
img_h[:,:,2] = img[:,:,2]*filter_h
```

```python
plt.imshow(img_h, cmap = 'hsv')
```

```python
filter_s = img_hsv[:,:,1] > 150
```

```python
img_s = img.copy()
```

```python
img_s[:,:,0] = img[:,:,0]*filter_h*filter_s
img_s[:,:,1] = img[:,:,1]*filter_h*filter_s
img_s[:,:,2] = img[:,:,2]*filter_h*filter_s
```

```python
plt.imshow(img_s, cmap = 'hsv')
```

```python
# For Extracting Yellow Pencil
# H = >25 <40
# S = > 200
# V = > 200

ch_h = img_hsv[:,:,0] > 25
ch_h1 = img_hsv[:,:,0] < 40
ch_s = img_hsv[:,:,1] > 200
ch_v = img_hsv[:,:,2] > 200

filter_y = ch_h*ch_h1*ch_s*ch_v
```

# Conclusion:
- To extract objects with specific color RGB Model is not used, instead HSV can be used.
- This is because Hue channel of HSV indicates chroma information of each pixel.
- Saturation channel of HSV model shows the amount by which the color is diluted.
- Value is the channel which indicates the overall brightness of the channel.
- It can be concluded that segmentation of object with a specific color is possible with HSV or similar models
